**INTRODUZIONE.**
Nella natura ci sono molti fenomeni che presentano delle *"oscillazioni"*: essendo lo studio della *natura*, la fisica tratta le *oscillazioni* - in particolare nel contesto della *fisica classica*. Si studiano ad esempio i *pendoli*, le *onde meccaniche* che si presentano in più forme, ed eccetera. Tuttavia per poter parlare delle oscillazioni, bisogna darne una breve definizione.

> **Definizione.** (Moto ondulatorio)
>
> Una quantità fisica $x:T\longrightarrow X$ (tradotto dal matematichese: una funzione al variare in tempo che ci dà la grandezza fisica in misura $X$) si dice *moto ondulatorio* se presenta l'equazione differenziale
> $$ \ddot x(t) + \omega^2 x(t)=0 $$

Notiamo che generalmente l'equazione differenziale viene risolta con la seguente formula:
$$ x(t)=A\cos(\omega t + \varphi) $$
con $A$ e $\varphi$ delle costanti che dipendono dalle *condizioni iniziali*.

Adesso ci concentriamo su un caso particolare del moto ondulatorio: il *pendolo*, osservato studiato in una maniera sistematica e rigorosa dallo scienziato Galileo Galilei; storicamente si narra questo studio scaturì nel 1581 nella Cattedrale di Pisa. Considerando la *dinamica* di un pendolo appeso su un filo lungo $L$, con un'angolazione di $\theta$ dalla verticale, si ha che la *sommatoria delle forze* è
$$ \left(\sum \vec{F}\right) \hat{i}' = ma = mg \sin(\theta) $$
Considerando che l'accelerazione $a$ rappresenta uno spostamento dell'angolo in $L \cdot \theta$, ricaviamo l'equazione differenziale
$$ L \ddot \theta = mg \sin \theta$$
Abbiamo qualcosa che ci *ricordi* l'equazione differenziale del moto associato alle ondulazioni: infatti, approssimando $\sin\theta \sim \theta$ (valido per piccoli angoli: infatti stiamo prendendo la formula di Taylor per il seno con $n=1$) ottenendo dunque il risultato
$$\boxed{ \omega = \sqrt{\frac{g}{L}} }$$
(ricordiamoci di questo risultato, che sarà l'importante per l'esperimento stesso!)
da cui ricaviamo il risultato finale
$$\boxed{\theta(t)=A \cos \left(\frac{g}{L}t + \varphi\right)}$$
Come ultima osservazione notiamo che possiamo giostrare il valore $\omega$ come
$$\omega^2 = \frac{g}{L} \implies \left(\frac{2\pi}{T}\right)^2 = \frac{g}{L} \implies \boxed{g=\frac{(2 \pi)^2 \cdot L}{T^2} }$$
che ci permette ricavare la costante $g$ da un semplice moto ondulatorio. $\blacksquare$

**Formule.**
> **Teorema.** (Formula della costante gravitazionale)
> 
> Prendendo il *moto semplice di un pendolo*, con raggio $L$, tempo di oscillazione $T$ vale che la costante gravitazionale $g$ può essere calcolata come segue nella formula.
> $$ \boxed{g=\frac{(2 \pi)^2 \cdot L}{T^2} } $$

$$
